**What is the purpose of a PivotTable in Excel?**

### **Answer:**

### **Question 4:**

**Explain the difference between Sorting and Filtering with an example from this dataset.**

### **Answer:**

**Sorting** means arranging the data in a specific order — either **small to large**, **large to small**, **A to Z**, or **Z to A**.
It changes **the order** of all the rows but **does not hide anything**.

**Filtering** means showing **only the rows you want to see** and hiding the rest.
It does **not change the order**, but it **hides unwanted data**.

###  **Example (Using this dataset: Product – Sales)**

| Product | Sales |
| ------- | ----- |
| Apple   | 2000  |
| Banana  | 1500  |
| Milk    | 3000  |
| Bread   | 1200  |

#### **Sorting Example:**

If you **sort Sales from largest to smallest**, the new order will be:

1. Milk – 3000
2. Apple – 2000
3. Banana – 1500
4. Bread – 1200

Nothing is hidden — only the order changed.

#### **Filtering Example:**

If you **filter Product = Apple**, then:

You will see only:

* Apple – 2000

Banana, Milk, and Bread will be **hidden**, not deleted.

## ✅ **Question 7:**

**Why are Dashboards important for decision-making in business analysis?**

### **Answer:**

* Dashboards give **all important data in one place**
* They show insights using **charts, KPIs, and visuals**
* Help managers take **quick and accurate decisions**
* Highlight **trends, patterns, and performance**
* Identify **problems and opportunities** early
* Reduce manual work and save **time**
* Make data easy to **understand, compare, and track**
